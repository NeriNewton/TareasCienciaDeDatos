# Canada Temperature Dataset

En este notebook desarrollo un análisis de datos exploratorio sobre el data set descargado de Kaggle: [https://www.kaggle.com/datasets/sarahquesnelle/canada-data](https://).

En este archivo se tienen los siguientes elementos:

* Year (Año en que se realizó la medición)

* Month (El mes de la medición)

* Stn_Name (Nombre de la localización de la estación de monitoreo del tiempo)

* Prov (Provincia de Canadá)

* Tm (Temperatura promedio durante un mes)

* S (Caída de nieve en cm por mes)

* P (Precipitación en mm por mes)



Con esto en mente es casi inmediato pensar en buscar indicios que sugieran que la temperatura ha aumentado conforme pasan los años. De esta manera la pregunta central de este análisis es:

***¿La temperatura promedio en Canadá ha aumentado en los últimos años?***

De esta pregunta podemos derivar algunas cuestiones más:

***¿La cantidad de nieve que cae se ve afectada por la temperatura?***

***¿Las precipitaciones se ven afectadas por los cambios de Temperatura?***

## Paqueterías

Primero es necesario importar las paqueterías y guardar en una variable el data frame:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = pd.read_csv('Canada_Temperature_Data.csv')
```

## Primera inspección



Primero es importante analizar las primeras filas de cada columna.

```python
data.head()
```

Ahora es necesario comprobar que las variables tengan el tipo adecuado. Para esto usaremos la función info:

```python
data.info()
```

Aquí se nota que existe una irregularidad con la variable P, pues si los datos están en unidades de mm, debemos esperar valores reales o incluso enteros, sin embargo, obtemos que P es del tipo objeto, esto puede implicar que existe texto dentro de los valores de la variable.

## Limpieza

Para eliminar el error con el tipo de variable que les corresponde a los datos de la columna P es necesario encontrar los valores que no son números reales, para esto primero es necesario encontrar los errores y verificar si se trata de un error regular o no.

```python
column_float = pd.to_numeric(data['P'], errors='coerce')

# Find the non-numeric values by checking for NaN values
non_numeric_values = data['P'][column_float.isna()]

# Print the non-numeric values
print(non_numeric_values)
```

Con esto es posible notar que los errores en P son provocados por la existencia de la cadena '#####' que se repite 101 veces, afortunadamente esta cantidad es mucho menor que la cantidad total de datos (1357283), por lo que podemos eliminar estos valores para convertir a P en una variable de tipo float64.

```python
data = data[data['P'] != '######']
data['P'] = data['P'].astype('float64')
```

## Descripción

Con los datos limpios, es momento de ver la estadística de los datos.

```python
data.describe().T
```

Ya que tenemos variables categóricas, es importante reconocer los valores únicos en estas.

```python
data['Prov'].unique()
```

```python
data['Stn_Name'].unique()
```

Y, como era de esperarse, obtuvimos los nombres de las provincias canadienses, así como los nombres de las regiones en las que se encuentran las estaciones meteorológicas. 

Ahora nos podemos permitir observar la distribución (en cantidad) de cada una de las variables numéricas.

```python
data.hist(bins=20, figsize=(10, 10))
plt.show()
```

Notamos que tenemos datos desde poc antes de 1920 hasta cerca de 2020, es decir, tenemos un registro mensual de datos de temperatura, precipitación y nieve a lo largo de 100 años.

Además, es efectivo utilizar una matriz de correlacción.

```python
corr = data.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.show()
```

De la matriz es posible observar que no hay ninguna correlación importante (además de las triviales), pero existe cierta correlación negativa entre los cm de nieve registrados y la temperatura, aunque esto puede sonar obvio, la correlación no es muy grande.

## Análisis

Para comenzar el análisis puede ser útil agrupar datos por año y temperatura, tomando la media, los máximos y los mínimos de cada año registrado, lo mismo para la precipitación y la nieve.

```python
max_temps = data.groupby('Year')['Tm'].max()
min_temps = data.groupby('Year')['Tm'].min()
mean_temps = data.groupby('Year')['Tm'].mean()
mean_p = data.groupby('Year')['P'].mean()
mean_s = data.groupby('Year')['S'].mean()
```

### Temperatura con datos totales

Con los datos agrupados observamos el comportamiento promedio de la temperatura media registrada cada año:

```python
plt.plot(mean_temps.index, mean_temps.values)

plt.xlabel('Año')
plt.ylabel('Temperatura Promedio (°C)')
plt.title('Año vs. Temperatura Promedio')

plt.show()
```

Lo mismo con los valores máximos.

```python
plt.plot(max_temps.index, max_temps.values)

plt.xlabel('Año')
plt.ylabel('Temperatura Máxima (°C)')
plt.title('Año vs. Temperatura Máxima')

plt.show()
```

Y para los mínimos.

```python
plt.plot(min_temps.index, min_temps.values)

plt.xlabel('Año')
plt.ylabel('Temperatura Mínima (°C)')
plt.title('Año vs. Temperatura Mínima')

plt.show()
```

De las tres gráficas anteriores es posible notar que existe una tendencia ligeramente decreciente (incluso neutra/constante) en las temperaturas desde 1920 hasta poco antes de 1980, año en que parece existir un punto de inflexión en la tendencia, marcando un crecimiento en los valores de las temperaturas a partir de este año.

### Temperatura con datos separados

Por lo anterior es necesario partir los datos en dos secciones, una que vaya desde 1917 hasta 1975, y otra desde 1971 hasta 2017.

```python
sub1 = data.loc[(data['Year'] >= 1917) & (data['Year'] <= 1970)]
sub2 = data.loc[(data['Year'] >= 1971) & (data['Year'] <= 2017)]


meTs1 = sub1.groupby('Year')['Tm'].mean()
meTs2 = sub2.groupby('Year')['Tm'].mean()
```

```python
plt.plot(meTs1.index, meTs1.values)

plt.xlabel('Año')
plt.ylabel('Temperatura Promedio (°C)')
plt.title('Temperatura Hasta 1970')

plt.show()
```

Es claro que la tendencia de la temperatura antes de 1970 es decreciente.

```python
plt.plot(meTs2.index, meTs2.values)

plt.xlabel('Año')
plt.ylabel('Temperatura Promedio (°C)')
plt.title('Temperatura Desde 1971')

plt.show()
```

Y que luego de 1970 se presenta una tendencia completamente creciente sin señales de cambio hasta 2017.

Sin embargo, este argumento es débil, para reforzar el resultado podemos haver uso de un box plot:

```python
plt.boxplot([meTs1,meTs2])

plt.title('Box Plot de Temperaturas Promedio')
plt.xlabel('Subconjuntos por tiempo')
plt.ylabel('Valores (°C)')

plt.show()
```

Es posible notar que cada uno de los valores (el quartil inferior, la media y el cuartil inferio) de la muestra 2 (correspondiente a los registros después de 1970) están por encima de los análogos de la muestra 1. Esto constituye una fuerte evidencia y forma un argumento más sólido para poder concluir que la temperatura promedio sí ha aumentado en Canadá.

```python
sns.violinplot([meTs1,meTs2])

plt.title('Violin Plot de Temperaturas Promedio' )
plt.xlabel('Subconjuntos por tiempo')
plt.ylabel('Valores(°C)')

plt.show()
```

Del mismo modo es posible revisar esta tendencia mediante la comparación de violin plots, en donde ahora es posible notar que la mayor densidad de valores de la segunda muestra se encuentra por encima de la primera, lo que de nuevo implica que la temperatura ha presentado un aumento desde 1970, además, la segunda muestra presenta una dispersión de datos más alta, lo que puede indicar que no solo la temperatura ha aumentado, sino que los hay cambios más notorios entre las temporadas frías y las temporadas cálidas.

### Precipitación

```python
plt.plot(mean_p.index, mean_p.values)

plt.xlabel('Año')
plt.ylabel('Precipitación Promedio (mm)')
plt.title('Año vs. Precipitación Promedio')

plt.show()
```

```python
meps1 = sub1.groupby('Year')['P'].mean()
meps2 = sub2.groupby('Year')['P'].mean()

plt.boxplot([meps1,meps2])

plt.title('Box Plot de Precipitación Promedio')
plt.xlabel('Subconjuntos de Precipitación')
plt.ylabel('Valores (mm)')

plt.show()
```

En este caso la lluvia parece obedecer una tendencia creciente desde el inicio del registro, lo que se refuerza con la comparación de box plots, sin embargo, no es posible concluir que el aumento de la temperatura esté correlacionado con el aumento en las lluvias, pues cuando la temperatura presenta una tendencia decreciente, la lluvia mantiene una tendencia creciente. Por lo que la temperatura y la precipitación registrada no parecen guardar una correlación.

### Nieve

```python
plt.plot(mean_s.index, mean_s.values)

plt.xlabel('Año')
plt.ylabel('Nivel de Nieve Promedio (cm)')
plt.title('Año vs. Nivel de Nieve Promedio')

plt.show()
```

```python
meSs1 = sub1.groupby('Year')['S'].mean()
meSs2 = sub2.groupby('Year')['S'].mean()

plt.boxplot([meSs1,meSs2])

plt.title('Box Plot de Nivel de Nieve Promedio')
plt.xlabel('Subconjuntos de Nieve')
plt.ylabel('Valores (cm)')

plt.show()
```

En este caso notamos que la tendencia es inversa a la presentada por la temperatura, sin embargo, este comportamiento es sutil. Al observar los boxplots podemos notar que la media del segundo conjunto de datos queda por debajo de la media del primer conjunto, sin embargo, los otros valores del conjunto dos se mantienen por encima de los análogos del primero, con lo que podemos asumir, siguiendo lo expuesto en la tabla de correlación, que el nivel de nieve promedio ha sido menor desde que se presentó el aumento de la temperatura en Canadá.

# Conclusiones

Con base en el desarrollo anterior es posible concluir que la temperatura promedio en Canadá ha presentado una tendencia creciente desde 1970, además, este comportamiento no presenta ningún indicador que anuncié un cambio de tendencia. A la vez, se han proporcionado algunas evidencias sobre el aumento en el nivel de precipitación registrado, sin embargo, este aumento parece no guardar una relación clara con el aumento de la temperatura, por lo que se recomienda obtener más datos que puedan relacionar estas dos variables en busca de una posible causa. Y, finalmente, se ha encontrado que el nivel de nieve registrado desde 1970 ha descendido levemente en comparación con los valores registrados antes de esta fecha, por lo que se concluye que la caída de nieve y el aumento de temperatura guardan una correlación inversa, siendo la temperatura la variable determinante.
